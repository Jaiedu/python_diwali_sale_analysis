```python
import numpy as np #Use for multi-dimensional arrays, matrices & mathematical functions
import pandas as pd #Use for for data manipulation, analysis, cleaning, & filtering etc
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline 
import seaborn as sns # create informative and attractive statistical graphics and visualizations
```

```python
df=pd.read_csv('Diwali Sales Data.csv', encoding='unicode_escape')
```

```python
df.shape #return row and columns
```

```python
df.head() #return 5 rows from top
```

```python
df.head(20) #return 20 rows from top
```

```python
#Data Cleaning
```

```python
df.info() #provide concise summary of the essential information about a DataFrame
```

```python
#drop blank columns / irrelevant columns
df.drop(['Status','unnamed1'], axis=1, inplace=True)
```

```python
df.info()
```

```python
#To check the null values, return True(Value Missing) or False(Value available)
pd.isnull(df)
```

```python
#To check the total number of values
pd.isnull(df).sum()
```

```python
#To drop the null values
df.dropna(inplace=True) #To save the value use (inplace=true)
```

```python
df.shape
```

```python
#Change Data Type
df['Amount']=df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
#To See the list of all columns
df.rename(columns={'Marital_Status':'Shaadi'})
```

```python
#To see the description of data in the Dataframe(count, mean,std etc)
df.describe()
```

```python
#To see the description of data for specific columns
df[['Age','Orders','Amount']].describe()
```

```python
#Exploratry Data Analysis EDA
```

```python
#Gender
```

```python
df.columns
```

```python
ax=sns.countplot(x='Gender',data=df)
```

```python
ax=sns.countplot(x='Gender',data=df) #With Labels
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Gender',y='Amount',data =sales_gen)
```

```python
#Conclusion: We can see that most of the buyers are females and purchasing power of females are greater than men
```

```python
#Age
ax=sns.countplot(x='Age Group', hue='Gender', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Total Amount Vs Age Group
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group', y='Amount', data= sales_age)
```

```python
#Conclusion: Most of the buyers are age group of between 26-35yrs female
```

```python
# State
```

```python
#Total number of orders from Top 10 states
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(15)
sns.set(rc={'figure.figsize':(20,3)})
sns.barplot(data=sales_state,x='State', y='Orders')
```

```python
#Total amount/sales from top 10 states
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(15)
sns.set(rc={'figure.figsize':(20,3)})
sns.barplot(data=sales_state,x='State', y='Amount')
```

```python
#Conclusion most of the orders are from uttar pradesh, Maharashtra and Karnatka.
```

```python
#Marital Status
ax=sns.countplot(x='Marital_Status', data=df)
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=sales_state,x='Marital_Status', y='Amount', hue='Gender')
```

```python
#Conclusion: Most of the buyers are married(Women) and they have high purchasing power
```

```python
#Occupation
```

```python
ax=sns.countplot(x='Occupation', data=df)
sns.set(rc={'figure.figsize':(25,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Purchasing Power
sales_state=df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=sales_state,x='Occupation', y='Amount')
```

```python
#Conclusion: Most of the buyers are working in IT, Aviation and Healthcare Sector
```

```python
#Products
ax=sns.countplot(x='Product_Category', data=df)
sns.set(rc={'figure.figsize':(25,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Purchasing Power
sales_state=df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=sales_state,x='Product_Category', y='Amount')
```

```python
#Conclusion: Most of the sold products are from Food, Footwear and Electronics Category
```
